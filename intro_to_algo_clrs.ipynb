```python
def insertion_sort_i(A):
	for j in range(1, len(A)):
		key = A[j]
		i = j - 1
		while i >= 0 and A[i] > key:
			A[i + 1] = A[i]
			i = i - 1
		A[i + 1] = key

def insertion_sort_d(A):
	for j in range(1, len(A)):
		key = A[j]
		i = j - 1
		while i >= 0 and A[i] < key:
			A[i + 1] = A[i]
			i = i - 1
		A[i + 1] = key


if __name__ == "__main__":
	# example 1: A = [5,2,4,6,1,3]
	# example 2: A = [31,41,59,26,41,58]

	A = [int(s) for s in input().split(",")]
	print("unsorted A: ", A)
	insertion_sort_i(A)
	print("increasing sorted A: ", A)
	insertion_sort_d(A)
	print("decreasing sorted A: ", A)
```

```python
def selection_sort(A):
    for i in range(0, len(A)):
        min = i
        for j in range(i+1, len(A)):
            # find the index of the ith smallest element
            if A[j] < A[min]:
                min = j
        # swamp A[min] and A[i]
        c = A[min]
        d = A[i]
        A[min] = d
        A[i] = c
        
if __name__ == "__main__":
    # example 1: A = [5,2,4,6,1,3]
    # example 2: A = [31,41,59,26,41,58]

    A = [int(s) for s in input().split(",")]
    print("unsorted A: ", A)
    selection_sort(A)
    print("sorted A: ", A)
```

```python
def merge_sort(A, p, q, r):
    
    n1 = q - p + 1
    n2 = r - q
    
    L = dict()
    R = dict()

    for i in range(1,n1+1):
        L[i] = A[p+i-1]

    for j in range(1,n2+1):
        R[j] = A[q+j]
    
    L[n1+1] = 1e1000 
    R[n2+1] = 1e1000

    i= 1
    j= 1
    for k in range(p, r+1):
        if L[i] <= R[j]:
            A[k] = L[i]
            i = i + 1
        else:
            A[k] = R[j]
            j = j + 1


if __name__ == "__main__":
    # Example A= [2,4,5,7,1,2,3,6] ; MERGE(A,9,12,16)
    A = [int(s) for s in input("Enter the values of the array comma seperated").split(",")]

    p = int(input("Enter the value of p"))
    q = int(input("Enter the value of q"))
    r = int(input("Enter the value of r"))

    A = {index:A[zero_index] for zero_index, index in enumerate(range(p,r+1))}
    
    print("unsorted A: ", list(A.values()))
    merge_sort(A, p, q, r)
    print("sorted A: ", list(A.values())) 
```
